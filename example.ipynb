```python
%matplotlib inline
import matplotlib.pyplot as plt
from squidAnalysis import loadSquidData
plt.rcParams.update({'font.size': 14, 'xtick.top': True, 'ytick.right': True, 'xtick.direction': 'in', 'ytick.direction': 'in'})
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 4), facecolor='w')

data =  loadSquidData("example_data_4.dat")
plt.plot(data['field'], data['moment']*1e3)

plt.xlabel("Applied field (mT)")
plt.ylabel("Moment (m-emu)")

plt.show()
```
